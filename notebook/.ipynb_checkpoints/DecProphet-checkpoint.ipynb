```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from decprophet import DecProphet
```

```python
pd.plotting.register_matplotlib_converters() # Volvemos a registrar los convertores.
plt.rcParams['figure.figsize'] = [16, 4]
```

```python
data = pd.read_csv("../data/tabla_generacion_energia.csv", parse_dates=["fecha"])
data.head()
```

```python
data = data.rename(columns={"fecha":"ds", "potencia":"y"})
data = data.drop(columns=["dir", "situacion"])
data = data.set_index("ds")
```

```python
dummy = pd.DataFrame()
dummy["sky1"] = (data.nub.isin([2,3]))
dummy["sky2"] = (data.nub.isin([4,5]))
dummy["sky3"] = (data.nub.isin([6,7]))
dummy["sky4"] = (data.nub.isin([8,9]))
dummy = dummy.astype("int64")
data = data.merge(dummy, right_index=True, left_index=True)
data.head()
```

```python
target = ["y"]
numerical = ["temp", "hum", "presion", "vel"]
categorical = ["nub"]
dummies = [col for col in data.columns if col not in target+numerical+categorical]
dummies
```

```python
train = data.loc["2019-06":"2019-07-25"].copy()
test = data.loc["2019-07-26":"2019-08-07"].copy()
```

```python
train.y.plot()
```

```python
test.y.plot()
```

```python
train = train.reset_index()
test = test.reset_index()
```

```python
model = DecProphet(exog=numerical+dummies, time_range=range(7,19))
```

```python
model.fit(train)
model.predict(test)
forecast = model.return_forecast()
```

```python
forecast.yhat.plot()
test.set_index("ds").y.plot()
```
